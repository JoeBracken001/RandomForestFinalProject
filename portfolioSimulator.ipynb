```python
import numpy as np
import pandas as pd
from sklearn import tree, ensemble
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
```

A Simple Portfolio Simulator
----------------------------------------

```python
df_results = pd.read_excel('PortfolioSimulator.xlsx',sheet_name='ForPython')
```

```python
df_DATE = df_results['Date']
date_time = pd.to_datetime(df_DATE)
df_strategyResults = df_results['Strategy']
df_BM = df_results['EW BM']

endValidDate=pd.Timestamp('2016-01-29 00:00:00')
df_strategyResults_OOS = df_strategyResults[df_DATE[df_DATE==endValidDate].index[0]:]
df_BM_OOS = df_BM[df_DATE[df_DATE==endValidDate].index[0]:]
date_time_OOS = date_time[df_DATE[df_DATE==endValidDate].index[0]:]
```

```python
cumRet_strategy = (df_strategyResults_OOS + 1).cumprod()
cumRet_BM = (df_BM_OOS + 1).cumprod()
plt.plot(date_time_OOS,cumRet_strategy, label = "Strategy")
plt.plot(date_time_OOS,cumRet_BM, label = "BM")
plt.legend(loc="upper left")
plt.show()
```
